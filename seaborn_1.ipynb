```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```
sns.set(style = 'darkgrid')
```

```
fmri = sns.load_dataset('fmri')
fmri.head()
```

```
sns.relplot(x = 'timepoint' , y = 'signal' , data = fmri , kind ='line' , ci = 'sd')
```

```
sns.relplot(x = 'timepoint' , y = 'signal' , data = fmri , estimator = None ,kind='line')
```

```
sns.relplot(x = 'timepoint' , y = 'signal' , data = fmri , kind ='line' ,hue = 'region' , style='event', ci = 'sd' , markers = True , dashes= False)
```

```
sns.relplot(x = 'timepoint' , y = 'signal' , data = fmri.query("event == 'cue'" ),kind = 'line' ,hue='event',style='event', ci = 'sd')
```

```
dots = sns.load_dataset('dots').query("align == 'dots'")
dots.head(5)
```

```
def sinplot(flip =1):
  x = np.linspace(0,14,100)
  for i in range(1,7):
    plt.plot(x , np.sin(x +i*0.5)*(7-i)*flip)
```
